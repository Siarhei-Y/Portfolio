```python
from time import sleep 
import pandas as pd

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver = webdriver.Chrome(executable_path='./chromedriver')
driver.get('https://www.fundoodata.com/advance_search_results.php?&new_industry_id[]=6&new_industry_id[]=125&new_industry_id[]=12&new_industry_id[]=29&new_industry_id[]=237&new_industry_id[]=26&new_industry_id[]=32&level_id=1&search_type=1')
sleep(3)
```

```python
data_list = []

active = True
#while we have a pagination

while active:
    
    #get list of companies on the page
    companies = driver.find_elements(By.XPATH, "//div[@class='heading']/a")
    
    #handle original page
    original_window = driver.current_window_handle
    
    #looping through companies on the page
    for company in range(len(companies)):
        #click on the company
        companies[company].click()
        WebDriverWait(driver, timeout=3).until(lambda d: d.find_elements(By.XPATH, "//div[@class='search-page-right-pannel']"))
        
        #handle clicked company
        driver.switch_to.window(driver.window_handles[-1])
        
        # get all necessary data on the clicked page ---------
        title = driver.find_element(By.XPATH, "//*[@id='main-container']/div/div[1]/div[2]/div[2]/div[1]").text
        phone_website = driver.find_element(By.XPATH, "//*[@id='main-container']/div/div[1]/div[2]/div[2]/div[3]").text
        industry = driver.find_element(By.XPATH, "//*[@id='main-container']/div/div[1]/div[2]/div[2]/div[4]/div[1]").text
        url = driver.current_url

        comp_item = {
            'title': title,
            'phone_website': phone_website,
            'industry': industry,
            'url': url
        }

        data_list.append(comp_item)
        
        # close a clicked page
        driver.close()
        
        #return a handle on the original page
        driver.switch_to.window(original_window)
        
        #refresh a company list
        companies = driver.find_elements(By.XPATH, "//div[@class='heading']/a")
        sleep(2)
    
    # scroll to top of the page
    driver.execute_script("window.scrollTo(0, 0)")
    
    #paginate
    try:
        pagination_btn = driver.find_element(By.XPATH, "(//div[@class='alphabet-search'])[2]/ul/li[7]/a")
        pagination_btn.click()
        sleep(3)
    except:
        active = False
        print('stopped')
```

```python
df = pd.DataFrame(data_list)
```
